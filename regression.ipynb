### Imports

```python
# Algemeen
import pandas as pd
import numpy as np
import math

# Hyper-parameter optimizing
from sklearn.model_selection import RandomizedSearchCV; from scipy.stats import uniform, randint
```

# Read Data in DataFrame

```python
# Training
train_set = pd.read_csv(data_path+"train.csv")
train_y = train_set['target']
train_X = train_set.drop(labels = ['target','id'], axis = 1)

# Testing
test_set = pd.read_csv(data_path+"test.csv")
test_X = test_set.drop(labels = 'id', axis = 1)
```

### Gradient boosting

```python
from sklearn.ensemble import GradientBoostingRegressor
reg = GradientBoostingRegressor(random_state=34)
```

Randomsearchcv

```python
search_dict = {'learning_rate': uniform(loc=0.000001, scale=0.999999),
                'n_estimators': randint(1, 25),
                'max_depth': randint(1, 25)} # [loc, loc + scale]
grad_boost= RandomizedSearchCV(reg, search_dict, random_state=0, n_iter=10,cv = 5, verbose=2,scoring = 'neg_mean_squared_error')
search = grad_boost.fit(train_X, train_y)
search.best_params_
search.best_estimator_
```

train best model

```python
best_grad_boost = search.best_estimator_.fit(train_X,train_y)
search.best_score_ # 0.01807
```

predict:

```python
test_y = best_grad_boost.predict(test_X)#lol
```

```python
def make_submission(test_y,test_set = test_set, naam = 'latest'):
    ingevuld_df = pd.DataFrame({'id': test_set['id'], 'target': test_y})
    ingevuld_df.to_csv(path_or_buf=data_path +'Submission_'+ naam + '.csv', sep=',',index = False)
```

```python
make_submission(test_set, test_y, naam = 'grad_boost')
```
